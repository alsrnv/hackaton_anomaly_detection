```python
df = pd.read_csv('http_metrics.csv' )
df.head()
```

```python
df.shape
```

```python
# Модель	Метрика	timestamp	value	is_anomaly
# 'Web Response', 'Throughput', 'Apdex', 'Error'

result_df = df[['ds',  'web_response_time']]
result_df.rename(columns={'ds': 'timestamp', 'web_response_time': 'value'}, inplace=True)
result_df['Модель'] = 'Etna'
result_df['Метрика'] = 'Web Response'
result_df[['Модель', 'Метрика', 'timestamp', 'value']]
```

```python
result_df.timestamp = pd.to_datetime(result_df.timestamp)
result_df.info()
```

```python
result_df
```

```python
result_df['Модель'] = 'Etna'
result_df['Метрика'] = 'Web Response'
result_df[['Модель', 'Метрика', 'timestamp', 'value']]
```

```python
# Преобразуем столбец 'ds' в формат datetime и переименуем его в 'timestamp'
result_df.rename(columns={'value': 'target'}, inplace=True)

# Добавляем столбец 'segment', указывающий на один сегмент
result_df['segment'] = 'segment_1'

```

```python
result_df
```

```python
# Создаем TSDataset с частотой 'T' (минуты)
ts = TSDataset(result_df[['timestamp', 'target', 'segment']], freq="T")

anomaly_dict = get_anomalies_density(ts, window_size=18, distance_coef=1, n_neighbors=4)

# Визуализируем аномалии
plot_anomalies(ts, anomaly_dict)
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
print(anomaly_dict.keys())
```

```python
# Преобразуем результат в DataFrame
anomalies = []
for segment, anomaly_list in anomaly_dict.items():
    for anomaly in anomaly_list:
        anomalies.append({
            'timestamp': anomaly,
            'segment': segment,
            'is_anomaly': True,
        })

anomalies_df = pd.DataFrame(anomalies)
```

```python
'Web Response', 'Throughput', 'Apdex', 'Error'
```

```python
# Модель	Метрика	timestamp	value	is_anomaly


result_df = result_df.merge(anomalies_df[['timestamp', 'is_anomaly']], how='left', on='timestamp')
result_df['is_anomaly'] = result_df['is_anomaly'].fillna(False)
result_df=result_df.drop(['segment'],axis=1)
result_df
```
